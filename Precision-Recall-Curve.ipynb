```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pickle.load(open("precisionRecallCurve.pkl", "rb"))
```

```python
print(data)
```

```python
ydata = np.empty(len(data))
xdata = np.empty(len(data))

for i in range(len(data)):
    ydata[i] = data[i][0]
    xdata[i] = data[i][1]

print(ydata)
print(xdata)
```

```python
maxind = np.argmax(ydata)
print(maxind)

plt.figure(figsize=(6,6))
plt.plot(xdata, ydata)
plt.plot(xdata[maxind], ydata[maxind], "rx")
plt.text(xdata[maxind]+0.05, ydata[maxind]-0.005, "Max precision of " + '{:.2f}'.format(ydata[maxind]) + 
         " at 5 recalled \nitems (recall " + '{:.2f}'.format(xdata[maxind]) + ")")
plt.ylabel("Precision")
plt.xlabel("Recall")
plt.savefig("PrecisionRecallCurve.png")
plt.show()
```
