```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pickle.load(open("precisionRecallCurve.pkl", "rb"))
```

```python
print(data)
```

```python
ydata = np.empty(len(data))
xdata = np.empty(len(data))

for i in range(len(data)):
    ydata[i] = data[i][0]
    xdata[i] = data[i][1]

print(ydata)
print(xdata)
```

```python
maxind = np.argmax(ydata)
print(maxind)

plt.plot(xdata, ydata)
plt.axvline(xdata[maxind], color="red", label="maximum precision at 5 recalled items")
plt.ylabel("Precision")
plt.xlabel("Recall")
plt.show()
```
